[Poppler Utils](https://linux.die.net/man/1/pdftohtml)

Poppler Utils is a popular option for extracting data from PDFs. It has the pdftohtml library that automagically extracts the coodinates of the objects in a PDF file. This works very well. You may still need to edit the results after the fact. You should estimate prior to extraction and edit the results because the output is usually just close to what one needs.

Also, labeling the data generally requires additional work. For instance, some of the schedule output groups the date with the first team that plays on that date. Visually, these are distinct. Creating your own bounding boxes for extraction allows you to be explicity and to align what you see with what you extract.
